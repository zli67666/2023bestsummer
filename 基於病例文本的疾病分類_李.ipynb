# 資料庫建立

```python
import pandas as pd
import glob

# x__data 存文件内容
x_data = []

# 文件路徑
for directory_path in ['/content/drive/MyDrive/Colab Notebooks/consensus/0/*',
                      '/content/drive/MyDrive/Colab Notebooks/consensus/1/*',
                      '/content/drive/MyDrive/Colab Notebooks/consensus/2/*',
                      '/content/drive/MyDrive/Colab Notebooks/consensus/3/*']:

    # 所有文件路徑
    file_all = glob.glob(directory_path)

    for i, file_path in enumerate(file_all):
        # 使用 open()  打開文件
        with open(file_path, 'r', encoding='utf-8') as f:
            x = f.read()  # 使用 f.read() 讀取文件内容

        # 文件内容添加到列表中
        x_data.append(x)
        #x_data=np.append(x_data, x)

        # x 包含了所有文件的文本内容
        print(f"文件路徑 {i+1}: {file_path}")
        print(x)
        print("=" * 50)  #分隔
```
